```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("e_ticaret_veri_seti_aykiri_degerler_duzenlendi.csv")
```

```python
df.shape
```

```python
sayisal_tipteki_sutunlar = df.select_dtypes(include = ["int64","float64"]).columns.tolist()
sayisal_tipteki_sutunlar
```

```python
analiz_sutunlari = ["birim_fiyat","toplam_tutar","teslimat_gunu","musteri_puani"]
df[analiz_sutunlari].describe()
```

```python
# Histogram
for sutun in analiz_sutunlari:
    plt.figure(figsize = (6,3))
    plt.hist(df[sutun],bins = 50)
    plt.title(sutun)
    plt.show()
```
