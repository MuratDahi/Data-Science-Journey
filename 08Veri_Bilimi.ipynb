# Grup Bazlı Analiz
## Amaç:
  1. Veri setini farklı kırılımlara ayırarak grup bazlı analiz yapma

## Adımlar:
  1. Veri seti yüklemek
  2. Sütunları belirleme
  3. Kategoriye göre temel analiz
  4. Şehre  göre temel analiz
  5. Müşteri tipine göre temel analiz
  6. Ödeme türüne göre temel analiz
  7. Birden fazla sütun ile grup bazlı analiz

```python
import pandas as pd
```

```python
df = pd.read_csv("e_ticaret_veri_seti_aykiri_degerler_duzenlendi.csv")
```

```python
df.shape
```

```python
grup_sutunlari = ["kategori","sehir","musteri_tipi","odeme_turu"]
sayisal_analiz_sutunlari = ["birim_fiyat","toplam_tutar","teslimat_gunu","musteri_puani"]
```

```python
# 3. Kategoriye göre temel analiz
kategori_analizi = df.groupby("kategori").agg(
    {
        "toplam_tutar" : ["count","mean","sum"],
        "birim_fiyat": "mean",
        "musteri_puani":"mean",
        "teslimat_gunu":"median"
    }
).round(2)
kategori_analizi
    
```

```python
# 4. şehre göre temel analiz
sehir_analizi = df.groupby("sehir").agg(
    {
        "toplam_tutar" : ["count","mean","sum"],
        "birim_fiyat": "mean",
        "musteri_puani":"mean",
        "teslimat_gunu":"median"
    }
).round(2)
sehir_analizi
```

```python
# 7. Birden fazla sütun ile veri analizi
kategori_musteri_analizi = df.groupby(["kategori","musteri_tipi"]).agg(
    {
      "toplam_tutar": ["count", "mean", "sum"],
      "birim_fiyat": "mean",
      "musteri_puani": "mean",
      "teslimat_gunu": "median"
    }
).round(1)
kategori_musteri_analizi
```
