```python
import pandas as pd
```

```python
# 1. veri setini yükle
dataFrame = pd.read_csv("e_ticaret_veri_seti_veri_tipleri_duzenlendi.csv")
dataFrame.shape
```

```python
dataFrame = dataFrame.drop_duplicates()
dataFrame.shape
```

```python
def metin_normalize_et(series):
    return (
        series.astype(str)
        .str.strip()
        .str.lower()
        .str.replace("i̇", "i", regex=False)
        .str.replace("ı", "i", regex=False)
        .str.replace("ö", "o", regex=False)
        .str.replace("ü", "u", regex=False)
        .str.replace("ç", "c", regex=False)
        .str.replace("ş", "s", regex=False)
        .str.replace("ğ", "g", regex=False)
    )
    
```

```python
dataFrame["sehir"] = metin_normalize_et(dataFrame["sehir"])
```

```python
dataFrame["sehir"].unique()
```

```python
sehir_duzeltme_haritasi = {
    "istanbul": "İstanbul",
    "ankara": "Ankara",
    "izmir": "İzmir",
    "antalya": "Antalya",
    "konya": "Konya",
    "bursa": "Bursa",
    "adana": "Adana",
    "gaziantep": "Gaziantep",
    "mersin": "Mersin",
    "kayseri": "Kayseri",
    "eskisehir": "Eskişehir",
    "samsun": "Samsun",
    "trabzon": "Trabzon",
    "kocaeli": "Kocaeli"
}
```

```python
dataFrame["sehir"] = dataFrame["sehir"].replace(sehir_duzeltme_haritasi)
dataFrame["sehir"].unique()
```

```python
dataFrame["sehir"].value_counts()
```

```python
dataFrame["kategori"] = metin_normalize_et(dataFrame["kategori"])
```

```python
kategori_duzeltme_haritasi = {
    "elektronik": "Elektronik",
    "kitap": "Kitap",
    "kozmetik": "Kozmetik",
    "kozmetık": "Kozmetik",
    "spor": "Spor",
    "ev yasam": "Ev Yaşam",
    "ev ve yasam": "Ev Yaşam",
    "ev&yasam": "Ev Yaşam",
    "ofis": "Ofis"
}
dataFrame["kategori"] = dataFrame["kategori"].replace(kategori_duzeltme_haritasi)
print("\nKategori sütunu düzeltildikten sonraki frekanslar:")
dataFrame["kategori"].value_counts()
```

```python
dataFrame["odeme_turu"] = metin_normalize_et(dataFrame["odeme_turu"])
odeme_turu_duzeltme_haritasi = {
    "kredi karti": "Kredi Kartı",
    "kart": "Kredi Kartı",
    "banka karti": "Banka Kartı",
    "dijital cuzdan": "Dijital Cüzdan",
    "kapida odeme": "Kapıda Ödeme",
    "havale": "Havale/EFT",
    "eft": "Havale/EFT",
    "havale/eft": "Havale/EFT"
}
dataFrame["odeme_turu"] = dataFrame["odeme_turu"].replace(odeme_turu_duzeltme_haritasi)
                                             
```

```python
print("\nÖdeme türü sütunu düzeltildikten sonraki frekanslar:")
dataFrame["odeme_turu"].value_counts()
```

```python
dataFrame["musteri_tipi"] = metin_normalize_et(dataFrame["musteri_tipi"])

musteri_tipi_duzeltme_haritasi = {
    "mevcut": "Mevcut",
    "yeni": "Yeni",
    "vip": "VIP"
}

dataFrame["musteri_tipi"] = dataFrame["musteri_tipi"].replace(musteri_tipi_duzeltme_haritasi)

print("\nMüşteri tipi sütunu düzeltildikten sonraki frekanslar:")
print(dataFrame["musteri_tipi"].value_counts())
```

```python
# =========================================================
# 10. Adım: Mantıksal olarak hatalı kayıtları inceleyelim
# Negatif ya da anlamsız değerler bazı kayıtlarda veri kalitesi problemi oluşturabilir.
# =========================================================

hatali_kayitlar = dataFrame[
    (dataFrame["adet"] <= 0) |
    (dataFrame["birim_fiyat"] <= 0) |
    (dataFrame["toplam_tutar"] <= 0) |
    (dataFrame["teslimat_gunu"] < 0) |
    (dataFrame["musteri_puani"] < 0) |
    (dataFrame["musteri_puani"] > 5)
]
```

```python
dataFrame = dataFrame[
    (dataFrame["adet"] > 0) &
    (dataFrame["birim_fiyat"] > 0) &
    (dataFrame["toplam_tutar"] > 0) &
    (dataFrame["teslimat_gunu"] >= 0) &
    (dataFrame["musteri_puani"] >= 0) &
    (dataFrame["musteri_puani"] <= 5)
]
```
