## Adımlar:
  1. Veri setini yükleme
  2. Pivot tablo için kullanılacak sütunların belirlenmesi
  3. Kategori ve müşteri tipine göre pivot tablo oluşturma
  4. Şehir ve ödeme türüne göre pivot tablo oluşturma
  5. Kategori ve ödeme türüne göre ortalama sipari tutarını inceleme
  6. Şehir ve müşteri tipine göre sipariş sayılarını inceleme
  

```python
import pandas as pd
```

```python
df = pd.read_csv("e_ticaret_veri_seti_aykiri_degerler_duzenlendi.csv")
df.shape
```

```python
# 2. Pivot tablo için kullanılacak sütunları belirle
pivot_satir_sutunlari = ["kategori","sehir"],
pivot_sutun_sutunlari = ["musteri_tipi","odeme_turu"]
deger_sutunlari = ["toplam_tutar","birim_fiyat"]
```

```python
# 3. Şehir ve ödeme türüne göre pivot tablo oluşturalım
pivot_kategori_musteri = pd.pivot_table(
    df,
    index = "sehir",
    columns = "odeme_turu",
    values = "toplam_tutar",
    aggfunc = "sum"
).round(1)
pivot_kategori_musteri
```

```python
# 4. Kategori ve müşteri tipine göre pivot tablo oluşturalım
pivot_kategori_musteri = pd.pivot_table(
    df,
    index="kategori",
    columns="musteri_tipi",
    values="toplam_tutar",
    aggfunc="mean"
).round(2)

print("\nKategori ve müşteri tipine göre ortalama toplam tutar pivot tablosu:")
pivot_kategori_musteri
```
