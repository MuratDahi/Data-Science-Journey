# Veri Tipleri ve Dönüşümler
### Amaç:
1. Veri setindeki sütunların veri tiplerini incele ve gerekli dönüşümleri gerçekleştir
  
### Adımlar:
1. Veri setini yüklemek
2. Veri tiplerini incele
3. Sayısal kategorik ve tarih sütunlarını belirle
4. Sipariş tarihi ni datetime sütununa dönüştür
5. Tarih sütunundan yeni özellikler üret
6. Kategorik sütunları kategori tipine dönüştür
7. Dönüşüm sonrası kontrol
8. Güncellenmiş veri setini kaydet.

```python
import pandas as pd
```

```python
# 1 veri setini yüklemek
dataFrame = pd.read_csv("e_ticaret_veri_seti_eksik_veriler_duzenlendi.csv")
```

```python
# 2 veri tiplerini incele
dataFrame.dtypes
```

```python
# 3. sayısal, kategorik ve tarih sutunlarını belirle
sayisal_sutunlar = dataFrame.select_dtypes(include = ["float64","int64"]).columns
kategorik_sutunlar = dataFrame.select_dtypes(include = ["object"]).columns
tarih_sutunlari = dataFrame["siparis_tarihi"]
```

```python
# 4. sipariş tarihini datetime tipine dönüştür
dataFrame["siparis_tarihi"] = pd.to_datetime(dataFrame["siparis_tarihi"])
dataFrame["siparis_tarihi"].dtypes
```

```python
# 5. tarih sütununda yeni özellikler türet
# yıl, ay, gün ve hafta günü bilgisini yazdır
dataFrame["siparis_yili"] = dataFrame["siparis_tarihi"].dt.year
dataFrame["siparis_ay"] = dataFrame["siparis_tarihi"].dt.month
dataFrame["siparis_gunu"] = dataFrame["siparis_tarihi"].dt.day
dataFrame["haftanin_gunu"] = dataFrame["siparis_tarihi"].dt.day_name()
dataFrame.head()
```

```python
dataFrame.columns
```

```python
# 6. kategorik sütunları kategori tipine dönüştür
kategorik_donusum_sutunlar = [
    "sehir",
    "kategori",
    "urun_adi",
    "odeme_turu",
    "musteri_tipi",
    "haftanin_gunu"
]

for sutun in kategorik_donusum_sutunlar:
    dataFrame[sutun] = dataFrame[sutun].astype("category")
```

```python
dataFrame.dtypes
```

```python
# 8. güncellenmiş veri setini kaydet
dataFrame.to_csv("e_ticaret_veri_seti_veri_tipleri_duzenlendi.csv",index = False, encoding = "utf-8-sig")
```
