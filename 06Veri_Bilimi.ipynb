# Kategorik Veri Analizi
## Amaç:
  1. Veri setindeki kategorik sütunları incelemek ve dağılımlawrını analiz etmek

## Adımlar:
  1. Veri setini yüklemek
  2. Kategorik sütunları belirle
  3. Kategorik sütunların genel bilgisini incele
  4. Her kategorik sütun için frekans analizi yapalım
  5. Her kategorik sütun için oran analizi yapalım
  6. En sık görülen kategorileri inceleme
  7. En az görüleni inceleme
  8. Detaylı analiz
  9. Count plot ile görselleştirme

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
# 1. veri setini yükle
dataFrame = pd.read_csv("e_ticaret_veri_seti_aykiri_degerler_duzenlendi.csv")
```

```python
# 2. kategorik sutunları belirle
kategorik_sutunlar = dataFrame.select_dtypes(include = ["object"]).columns.tolist()
kategorik_sutunlar
```

```python
# 4. frekans analizi
for sutun in kategorik_sutunlar:
    print(f"{sutun}")
    print(dataFrame[sutun].value_counts())
    print()
```

```python
# 5. oran analizi
for sutun in kategorik_sutunlar:
    print(f"{sutun}")
    print((dataFrame[sutun].value_counts(normalize = True) * 100).round(1))
    print()
```

```python
# 9. count plot ile görselleştirme
detayli_analiz_sutunlari = ["sehir","kategori","odeme_turu","musteri_tipi"]
for sutun in detayli_analiz_sutunlari:
    plt.figure(figsize= (6,3))
    sns.countplot(data = dataFrame, x = sutun, order = dataFrame[sutun].value_counts().index)
    plt.xticks(rotation = 45)
    plt.show()
    
```
