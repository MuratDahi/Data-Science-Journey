```python
# 4. eksik veri içeren satırları inceleme
# NaN sayısal sutünlardaki boşluk, none nesne türündeki(object) boşluk isnull() ikisinide alır.
eksik_veri_olan_satirlar = dataFrame[dataFrame.isnull().any(axis = 1)]
```

```python
eksik_veri_olan_satirlar.head()
```

```python
# 5. Eksik verileri içeren yerleri sil
df_silinen = dataFrame.dropna()
df_silinen.shape
```

```python
df_silinen
```

```python
# 6. sayısal eksik verileri ortalama
df_ortalama = dataFrame.copy()
sayisal_sutunlar = dataFrame.select_dtypes(include = ["float64","int64"]).columns
for sutun in sayisal_sutunlar:
    df_ortalama[sutun] = df_ortalama[sutun].fillna(df_ortalama[sutun].mean())
df_ortalama[sayisal_sutunlar].isnull().sum()
```

```python
# 7. sayısal eksik verileri median
df_ortalama = dataFrame.copy()
sayisal_sutunlar = df_ortalama.select_dtypes(include=["float64", "int64"]).columns

for sutun in sayisal_sutunlar:
  df_ortalama[sutun] = df_ortalama[sutun].fillna(df_ortalama[sutun].median())
df_ortalama[sayisal_sutunlar].isnull().sum()
```

```python
# 8. sayısal eksik verileri mod
df_mod = dataFrame.copy()
sayisal_sutunlar = df_mod.select_dtypes(include=["float64", "int64"]).columns

for sutun in sayisal_sutunlar:
    en_cok_tekrar_eden = df_mod[sutun].mode()[0]
    df_mod[sutun] = df_mod[sutun].fillna(en_cok_tekrar_eden)
df_ortalama[sayisal_sutunlar].isnull().sum()
```

```python
# 9. kategorik eskik verileri mod ile doldurma
df_mod_kategorik = dataFrame.copy()
kategorik_sutunlar = df_mod_kategorik.select_dtypes(include = ["object"]).columns
for sutun in kategorik_sutunlar:
    df_mod_kategorik[sutun] = df_mod_kategorik[sutun].fillna(df_mod_kategorik[sutun].mode()[0])
df_mod_kategorik[kategorik_sutunlar].isnull().sum()
```

```python
# 10. kategorik eksik verileri sabit bir değer ile doldurma
df_sabit_kategorik = dataFrame.copy()
for sutun in kategorik_sutunlar:
    df_sabit_kategorik[sutun] = df_sabit_kategorik[sutun].fillna("bilinmiyor")
df_sabit_kategorik[kategorik_sutunlar].isnull().sum()
```

```python
df_guncel = dataFrame.copy()

for sutun in kategorik_sutunlar:
    df_guncel[sutun].fillna(df_guncel[sutun].mode()[0])

for sutun in sayisal_sutunlar:
    df_guncel[sutun].fillna(df_guncel[sutun].median())
```

```python
df_guncel.to_csv("e_ticaret_veri_seti_eksik_veriler_duzenlendi.csv",index = False, encoding = "utf-8-sig")
```
